```python
#GENI CHATBOT
class GeniBot:
  def __init__(self):
    self.client=genai.Client(api_key=API_KEY)
    self.history=[]

  def chat(self,user_message:str)->str:
    self.history.append({"role":"user","parts":[{"text":user_message.strip()}]})

    response=self.client.models.generate_content(
        model=MODEL,
        contents=self.history,
        config=genai.types.GenerateContentConfig(
            system_instruction=SYSTEM_PROMPT,
            max_output_tokens=600,
        ),
    )

    reply=response.text
    self.history.append({"role":"model","parts":[{"text":reply}]})
    return reply

  def reset(self):
    self.history=[]
    print("\n]Conversation Cleared]\n")
```

```python
#Main Function
print("Welcome to LLM Hospital.\n")

bot=GeniBot()

while True:
  user_input=input("You: ").strip()
  if not user_input:
    continue
  if user_input.lower() in {"quit","exit","bye"}:
    print("Geni: Thank you for reaching out to LLM Hospital.\n Take care and stay healthy. Goodbye!")
    break
  if user_input.lower()=="reset":
    bot.reset()
    continue

  try:
    reply=bot.chat(user_input)
    print(f"\nGeni: {reply}\n")
  except Exception as e:
    print(f"\n[Error {e}\n]")
    print("Check your API key at https://aistudio.google.com\n")

```
